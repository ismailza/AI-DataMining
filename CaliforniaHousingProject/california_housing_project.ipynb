#### Manage missing values with Scikit-learn

```python
from sklearn.impute import SimpleImputer
```

```python
imputer = SimpleImputer(strategy = "median")
```

```python
data_num = data.drop("ocean_proximity", axis = 1)
```

```python
imputer.fit(data_num)
```

```python
imputer.statistics_
```

```python
data_num.median().values
```

```python
data_array = imputer.transform(data_num)
```

```python
data_array
```

```python
data_tr = pd.DataFrame(data_array, columns = data_num.columns, index = data_num.index)
```
